# Creating Tables

```python
import pandas as pd
authors_df = pd.read_csv('data/authors.csv')
books_df = pd.read_csv('data/books.csv')
users_df = pd.read_csv('data/users.csv')
fav_df = pd.read_csv('data/favorites.csv')

print("Authors")
display(authors_df)
print("===================\n Books")
display(books_df)
print("===================\n Users")
display(users_df)
print("===================\n Favorites")
display(fav_df)
```

```python
authors_df.to_sql('authors', engine, if_exists = 'replace')
q = """SELECT * FROM authors LIMIT 5;"""
pd.read_sql(q, engine)
```

```python
books_df.to_sql('books', engine, if_exists = 'replace')
q = """SELECT * FROM books LIMIT 6;"""
pd.read_sql(q, engine)
```

```python
users_df.to_sql('users', engine, if_exists = 'replace')
q = """SELECT * FROM users LIMIT 3;"""
pd.read_sql(q, engine)
```

```python
fav_df.to_sql('favorites', engine, if_exists = 'replace')
q = """SELECT * FROM favorites LIMIT 7;"""
pd.read_sql(q, engine)
```

```python
q = """SELECT books.title, favorites.user_id
FROM books
JOIN favorites ON books.id = favorites.book_id
WHERE favorites.user_id =
    (SELECT users.id FROM users WHERE (users.last_name = "Doe" AND users.first_name = "John"))"""
pd.read_sql(q, engine)
```
